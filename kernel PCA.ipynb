```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler

from sklearn import datasets  
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

x,y = datasets.make_circles(n_samples=1000,
                            random_state=123,
                            noise=0.1,
                            factor=0.2)

plt.scatter(x[y==0,0],x[y==0,1],
            c='blue',marker='o',alpha=0.5)
plt.scatter(x[y==1,0],x[y==1,1],
            c='red',marker='x',alpha=0.5)

plt.show()
```

PCA無法解決非線性分離資料

```python
from sklearn.decomposition import PCA

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))
pca =PCA(n_components=2)
x_pca = pca.fit_transform(x)

p = pca.explained_variance_ratio_
ax[0].scatter(x_pca[y==0,0],x_pca[y==0,1],
            c='blue',marker='o',alpha=0.5)
ax[0].scatter(x_pca[y==1,0],x_pca[y==1,1],
            c='red',marker='x',alpha=0.5)


ax[1].scatter(x_pca[y==0,0],np.zeros((500,1))+0.02,
            c='blue',marker='o',alpha=0.5)
ax[1].scatter(x_pca[y==1,0],np.zeros((500,1))-0.02,
            c='red',marker='x',alpha=0.5)

ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-1, 1])
ax[1].set_yticks([])
ax[1].set_xlabel('PC1')


plt.tight_layout
plt.show()


```

利用核PCA解決非線性分離

```python
from sklearn.decomposition import KernelPCA

x,y =datasets.make_moons(n_samples=100,random_state=123)

x_kpca = KernelPCA(n_components=2,kernel='rbf',gamma=15)

kpca = x_kpca.fit_transform(x)

plt.scatter(kpca[y==0,0],kpca[y==0,1],
            c='red',marker='o',alpha=0.5)
plt.scatter(kpca[y==1,0],kpca[y==1,1],
            c='blue',marker='x',alpha=0.5)

```
